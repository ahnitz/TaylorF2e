```python
from pycbc.waveform import get_fd_waveform
```

```python
hp, hc = get_fd_waveform(approximant="TaylorF2e",
                         mass1=1.4, mass2=1.4, 
                         f_lower=20.0, delta_f=1.0/128, 
                         distance=100, 
                         eccentricity=1e-7, 
                         inclination= 0.0,
                         coa_phase = 0, 
                         long_asc_nodes=0)
```

```python
import pylab
pylab.plot(hc.sample_frequencies, hc)
#pylab.plot(hp.sample_frequencies, hp)
pylab.xlim(30, 250)
pylab.xscale('log')
pylab.show()
```

```python
hp2, hc2 = get_fd_waveform(approximant="TaylorF2",
                         mass1=1.4, mass2=1.4, 
                         f_lower=20.0, delta_f=1.0/128, 
                         distance=100, 
                         eccentricity=0, 
                         inclination= 0,
                         coa_phase = 0, 
                         long_asc_nodes=0, phase_order=6)

import pylab
pylab.plot(hc.sample_frequencies, hc, label='eccentric')
pylab.plot(hc2.sample_frequencies, hc2, label='normal')
pylab.legend()
pylab.xlim(200, 220)
pylab.xscale('log')
pylab.show()
```

```python
from pycbc.filter import match
```

```python
mlen = max(len(hp), len(hp2))
hp.resize(mlen)
hp2.resize(mlen)
match(hp, hp2)
```
