```python
from pycbc.waveform import get_fd_waveform
```

```python
hp, hc = get_fd_waveform(approximant="TaylorF2e",
                         mass1=1.4, mass2=1.4, 
                         f_lower=40.0, delta_f=1.0/128, 
                         distance=100, 
                         eccentricity=1e-5, 
                         inclination= 0,
                         coa_phase = 1.7, 
                         long_asc_nodes=0)
```

```python
import pylab
pylab.plot(hc.sample_frequencies, hc)
#pylab.plot(hp.sample_frequencies, hp)
pylab.xlim(30, 250)
pylab.xscale('log')
pylab.show()
```

```python
hp2, hc2 = get_fd_waveform(approximant="TaylorF2",
                         mass1=1.4, mass2=1.4, 
                         f_lower=40.0, delta_f=1.0/128, 
                         distance=100, 
                         eccentricity=0, 
                         inclination= 0,
                         coa_phase = 1.7, 
                         long_asc_nodes=0)

import pylab
pylab.plot(hc.sample_frequencies, hc)
pylab.plot(hc2.sample_frequencies, hc2)
pylab.xlim(240, 250)
pylab.xscale('log')
pylab.show()
```
